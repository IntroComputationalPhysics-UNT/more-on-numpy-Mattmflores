## Operations: *, @, and np.matmul / np.dot

Elementwise and matrix multiplication are treated differently in Numpy.

### Elementwise multiplication (`*`)

```
A = np.array([[1,2],[3,4]])
B = np.array([[10,20],[30,40]])
A * B
```

### Matrix multiplication (`@`)

```
A @ B
```

### Function equivalents for matrix multiplication

```
np.matmul(A, B), np.dot(A, B)
```

```
# COMPLETE THESE TASKS
# 1. look into the documentation for `np.matmul`, `np.dot`, `@`, and `*` and explain their difference(s)
# 2. create an example/examples to explain their differences

# *        → elementwise multiplication (Hadamard product)
# @        → matrix multiplication (A @ B)
# np.matmul → same as @ but works with batches of matrices
# np.dot    → for 2D arrays it's the same as @, but for 1D arrays it gives the inner product

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[10, 20],
              [30, 40]])

# Elementwise multiplication
print("A * B =\n", A * B)

# Matrix multiplication
print("\nA @ B =\n", A @ B)

# np.matmul
print("\nnp.matmul(A, B) =\n", np.matmul(A, B))

# np.dot (same as @ for 2D)
print("\nnp.dot(A, B) =\n", np.dot(A, B))

# 1D example to show np.dot difference
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

print("\nv * w =", v * w)        # elementwise
print("np.dot(v, w) =", np.dot(v, w))  # inner product
print("v @ w =", v @ w)                  # same as dot for 1D

```

```
# COMPLETE THIS TASK
# create a vectorized function that calculates the total energy (T+U; kinetic + potential energy) of a harmonic oscillator
def harmonic_energy(m, k, x, v):
    """
    Vectorized total energy of a harmonic oscillator.
    m, k: scalars
    x, v: arrays or scalars (broadcasted)
    """
    x = np.asarray(x)
    v = np.asarray(v)
    T = 0.5 * m * v**2
    U = 0.5 * k * x**2
    return T + U

```

## Eigenvalues and Eigenvectors

For a matrix $M$, eigenvalues $\lambda$ and eigenvectors $v$ satisfy:
$$ M v = \lambda v $$

In physics, these often realte to characteristic and measurable physical quantities:
* Oscillators → $\lambda$ relates to squared frequencies
* Quantum mechanics → $\lambda$ could be the energy eigenstates
* Rotations → $\lambda$ could define a principal axes

### Here is an example matrix

```
K = np.array([[ 1,-1, 0]
             ,[-1, 2,-1]
             ,[ 0,-1, 1] ])
vals, vecs = np.linalg.eigh(K)
vals, vecs
```

```
# COMPLETE THIS TASK
# In the above cell, which eigenvector is associated with which eigenvalue?

# The i-th column vecs[:, i] is the eigenvector associated
# with eigenvalue vals[i].
# By definition of np.linalg.eigh:
# vals[0] ↔ vecs[:, 0]
# vals[1] ↔ vecs[:, 1]
# vals[2] ↔ vecs[:, 2]

for i, lam in enumerate(vals):
    print(f"Eigenvalue {i}: {lam}")
    print("  Eigenvector:", vecs[:, i])
```

## Quadratic Form: Diatomic Molecule Model

We can model a diatomic molecule (O$_2$ or N$_2$, for example) as two masses connected by a spring. Motion is restricted to the
bond axis.

The (quadratic) potential energy is: `V = 0.5 * x.T @ K @ x ` where `x = (x1, x2)`, the coordinates of the two atoms.

### Constructing a stiffness/force-constant matrix

```
k = 2450 # the force constant in an appropriate [energy]/[length] units
K = np.array([[ k,-k]
             ,[-k, k]])
K
```

### Diagonalize to find frequencies and normal mode displacement patterns

```
vals, vecs = np.linalg.eigh(K)
vals, vecs
```

### Vibrational frequencies (ignoring mass)

```
omega = np.sqrt(vals)
omega
```

### Unitary transformations with eigenvector matrices

If the matrix is not singular, we can find eigenvalues and eigenvectors. The matrix of eigenvectors $U$ forms a unitary transformation.

```
# COMPLETE THESE TASKS
# Show that U@U.T and U.T@U are 2x2 identity matrices
# Note that normally we would need to use the `.conj().T` attribute to find the conjugate transpose matric, but our matrix is real.
U = vecs  # eigenvector matrix

print("U @ U.T =")
print(U @ U.T)

print("\nU.T @ U =")
print(U.T @ U)

# For an orthonormal eigenvector matrix U, both expressions
# should be (numerically) equal to the 2x2 identity matrix:
print("\nCheck closeness to identity:")
print(np.allclose(U @ U.T, np.eye(2)))
print(np.allclose(U.T @ U, np.eye(2)))
```

```
# COMPLETE THIS TASK
# From the eigenvalue equation M v = λ v, it follows that for U = [v1, v2], then M U = λ U.
# Because λ is diagonal and U.T@U = np.eye(2), it follows that λ = U.T @ K @ U.
# Confirm this behavior.
# K, vals, vecs already defined above
U = vecs
Lambda = np.diag(vals)        # diagonal matrix of eigenvalues

left = U.T @ K @ U            # should equal Lambda

print("Lambda =")
print(Lambda)
print("\nU.T @ K @ U =")
print(left)

print("\nCheck equality:")
print(np.allclose(left, Lambda))
```

```
# COMPLETE THIS TASK
# Use the eigenvectors to eigenvalues to interpret the two modes in this model of a diatomic molecule

# Mode 1: eigenvalue ≈ 0, eigenvector ~ (1, 1)
#   Both atoms move in phase by the same amount.
#   This is a center-of-mass motion,
#   so the bond length does not change and there is effectively no restoring force.

# Mode 2: eigenvalue ≈ 2k, eigenvector ~ (1, -1)
#   The two atoms move in opposite directions.
#   This changes the bond length, so it corresponds to the
#   stretching of the diatomic molecule.

for i, lam in enumerate(vals):
    print(f"Mode {i}: eigenvalue = {lam}, omega = {np.sqrt(lam)}")
    print("  eigenvector =", vecs[:, i])
```
